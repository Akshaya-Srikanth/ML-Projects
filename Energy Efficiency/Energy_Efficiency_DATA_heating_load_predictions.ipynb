```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('ENB2012_data.csv')
data.dropna(how='all') #remove NaN values from the data set
datanew= data.dropna(axis=1, how='all') #first deletes all the non useful columns from the dataset
data_processed=datanew.dropna(how='all') #then deletes all the unuseful rows from the data set
print(data_processed)
#lets try re-naming all the headers columns
data_finalized=data_processed.rename(columns={"X1": "Relative Compactness", "X2": "Surface Area","X3": "Wall Area","X4": "Roof Area","X5": "Overall Height","X6": "Orientation","X7": "Glazing Area","X8": "Glazing Area Distribution","Y1": "Heating Load","Y2":"Cooling Load"})
data_finalized.head(10) #lets me look at random samples within the dataset
```

```python
x=data_finalized.iloc[:,:-2].values
y=heating_load_y=data_finalized.iloc[:,8:9].values
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
#very good value  r square value
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
