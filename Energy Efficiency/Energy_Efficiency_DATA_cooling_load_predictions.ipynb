# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import Data Set and Clean the dataset

```python
data=pd.read_csv('/content/sample_data/ENB2012_data.csv')
data.dropna(how='all') #remove NaN values from the data set
datanew= data.dropna(axis=1, how='all') #first deletes all the non useful columns from the dataset
data_processed=datanew.dropna(how='all') #then deletes all the unuseful rows from the data set
print(data_processed)
#lets try re-naming all the headers columns
data_finalized=data_processed.rename(columns={"X1": "Relative Compactness", "X2": "Surface Area","X3": "Wall Area","X4": "Roof Area","X5": "Overall Height","X6": "Orientation","X7": "Glazing Area","X8": "Glazing Area Distribution","Y1": "Heating Load","Y2":"Cooling Load"})
data_finalized.head(10) #lets me look at random samples within the dataset
```

# Define x and y variables

```python
x=data_finalized.iloc[:,:-2].values
y=data_finalized.iloc[:,9:10].values
print(y)
```

```python
#y = y.reshape(len(y),1) #do this only for feature scaling for Support Vector Regressor mode
```

# Split data into training and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

# Run this code for applying random forest regression

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# R^2 value- random forest regression model

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
#very good value  r square value
```

# Decision Tree Regression 

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# R^2- Decision tree regression

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

# Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree = 8)
x_poly = poly_reg.fit_transform(x_train)
regressor = LinearRegression()
regressor.fit(x_poly, y_train)
```

```python
y_pred = regressor.predict(poly_reg.transform(x_test))
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# R^2- Polynomial Regression

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

# Multiple Regression Model

```python

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# R^2-Multiple Regression Model

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
